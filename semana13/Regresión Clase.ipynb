```python
#Boxplots
fig,axs=plt.subplots(ncols=5,nrows=3,figsize=(20,10))
index=0
axs=axs.flatten()
for k,v in df.items():
    sns.boxplot(y=k,data=df,ax=axs[index])
    index+=1
plt.tight_layout(pad=0.4, w_pad=0.5,h_pad=5)
```

```python
# Porcentaje de outliers por columna
for k,v in df.items():
    q1=v.quantile(0.25)
    q3=v.quantile(0.75)
    dq=q3-q1
    v_col=v[(v<=q1-1.5*dq)|(v>=q3+1.5*dq)]
    perc=np.shape(v_col)[0]*100/np.shape(df)[0]
    print(k+"-"+str(perc))
```

```python
# Quitar las observaciones cuyo MEDV sea un outlier
MEDV_q3= df["MEDV"].quantile(0.75)
MEDV_q1= df["MEDV"].quantile(0.25)
MEDV_D=MEDV_q3-MEDV_q1
big_sup=MEDV_q3+1.5*MEDV_D
big_inf=MEDV_q1-1.5*MEDV_D
df=df[~(df["MEDV"]>=big_sup)]
df=df[~(df["MEDV"]<=big_inf)]
l=list(df.corr()[df.corr().abs()["MEDV"]>0.5].index)
df=df[l]
```

### Regresión lineal simple

Variable independiente: RM

Variable dependiente: MEDV

```python
# Mostrar la gráfica de regresión
plt.figure(figsize=(18,8))
ax=sns.regplot(x=df["RM"],y=df["MEDV"],marker="o")
```

```python
# Modelo de regresión
from sklearn.linear_model import LinearRegression

# Instancia del modelo
reg= LinearRegression()
```

```python
# Crear los datos X, Y y dividirlos en entrenamiento y prueba
# Importar la librería
from sklearn.model_selection import train_test_split
# Crear los datos X,Y
X=df["RM"].to_numpy().reshape(466,1)
Y=df["MEDV"]

# División en entrenamiento y prueba
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3)
```

```python
# Entrenar el modelo
reg.fit(Xtrain,Ytrain)

# Predecir los resultados para el conjunto de prueba
Ypred=reg.predict(Xtest)
```

```python
# Importar las métricas
from sklearn.metrics import r2_score,mean_squared_error

# Métricas
# Coeficiente de determinación
r2=r2_score(Ytest,Ypred)
mse= mean_squared_error(Ytest,Ypred)
print(r2,mse,mse**0.5)
```

```python
# Coeficientes del modelo
print(reg.coef_)
print(reg.intercept_)
```

### Regresión lineal múltiple

```python
# Crear los datos X,Y
X=df[df.columns[:-1]]
Y=df[df.columns[-1]]
# División en entrenamiento y prueba
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3)
```

```python
# Instanciar el modelo

reg= LinearRegression()
# Entrenar el modelo
reg.fit(Xtrain,Ytrain)

# Predecir los resultados para el conjunto de prueba
Ypred=reg.predict(Xtest)
```

```python
# Métricas
# Coeficiente de determinación

r2=r2_score(Ytest,Ypred)
mse= mean_squared_error(Ytest,Ypred)
print(r2,mse,mse**0.5)
```

```python
# Coeficientes del modelo
print(reg.coef_)
print(reg.intercept_)
```

### Regresión por árbol de decisión

```python
# Importar la librería
from sklearn.tree import DecisionTreeRegressor
```

```python
# Regresión simple (MR vs. MEDV)
# Instancia del modelo
dtr=DecisionTreeRegressor()
```

```python
# Entrenar el modelo
dtr.fit(Xtrain,Ytrain)
```

```python
# Predecir
Ypred =dtr.predict(Xtest)
```

```python
# Métricas
# Coeficiente de determinación
r2=r2_score(Ytest,Ypred)
mse= mean_squared_error(Ytest,Ypred)
print(r2,mse,mse**0.5)
```

```python
# Importancia de cada característica

dtr.feature_importances_
```

¿Porqué para el modelo no lineal pueden dar valores negativos de $r^2$?

Para el caso general se calcula como:

$$\rho^2=1-\frac{\sigma^2_r}{\sigma^2}$$

Siendo $\sigma^2_r$ la varianza residual y $\sigma^2$ la varianza de la variable dependiente.
