```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,4*np.pi,0.1)
```

```python
y = np.sin(x)
```

```python
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("x(radians)")
plt.ylabel("sin(x)/cos(x)")
plt.legend(["sin","cos"])
plt.title("seno y coseno con numpy")
plt.show()
```
