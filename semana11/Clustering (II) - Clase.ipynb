# Clustering o agrupamiento (II)

```python
# Importar las librerías básicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.dpi'] = 120
```

## Dataset artificial
*Lunas*

```python
# Librería para el dataset
from sklearn.datasets import make_moons

# Semilla de los aleatorios
np.random.seed(1235)

# Generación de dos semicírculos entrelazados
X, y = make_moons(n_samples = 200,
                 noise = 0.05)
```

```python
plt.scatter(X[:,0],X[:,1],s=5)
```

```python
# Importar librería
from sklearn.cluster import KMeans

# Instanciar el modelo
km = KMeans(n_clusters=8,init="random",n_init=10,max_iter=300,tol=1e-4)

# Predicción de los clústeres
y_km =km.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1], s=50, c='lightgreen', marker='s', edgecolor='black',label='cluster 1')
plt.scatter(X[y_km == 1, 0], X[y_km == 1, 1], s=50, c='orange', marker='o', edgecolor='black', label='cluster 2')
plt.scatter(X[y_km == 2, 0], X[y_km == 2, 1], s=50, c='red', marker='v', edgecolor='black',label='cluster 3')
plt.scatter(X[y_km == 3, 0], X[y_km == 3,1], s=50, c='green', marker='v', edgecolor='black',label='cluster 4')
plt.scatter(X[y_km == 4, 0], X[y_km ==4,1], s=50, c='blue', marker='v', edgecolor='black',label='cluster 5')
plt.scatter(X[y_km == 5, 0], X[y_km ==5,1], s=50, c='lightblue', marker='v', edgecolor='black',label='cluster 6')
plt.scatter(X[y_km == 6, 0], X[y_km ==6,1], s=50, c='yellow', marker='v', edgecolor='black',label='cluster 7')
plt.scatter(X[y_km == 7, 0], X[y_km ==7,1], s=50, c='white', marker='v', edgecolor='black',label='cluster 8')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, marker='*', c='red', edgecolor='black', label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
dist=[]
for i in range(1,20):
    kms=KMeans(n_clusters=i,init="k-means++",n_init=10,max_iter=100)
    kms.fit(X)
    dist.append(kms.inertia_)
plt.plot(range(1,20),dist,marker="o")
```

```python
# Importar librería
from sklearn.cluster import DBSCAN

# Instanciar el modelo

db = DBSCAN(eps=0.2,min_samples=5,metric="euclidean")

# Predecir los clústeres

y_db =db.fit_predict(X)
y_db
```

```python
plt.scatter(X[y_db == 0, 0], X[y_db == 0, 1], s=50, c='red', marker='o', edgecolor='black',label='cluster 1')
plt.scatter(X[y_db == 1, 0], X[y_db == 1, 1], s=50, c='blue', marker='o', edgecolor='black', label='cluster 2')
plt.legend(scatterpoints=1, loc = 0)
plt.grid()
plt.show()
```
